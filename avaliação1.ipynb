Se X e N(1000; 20), produza utilizando um software adequado (Python, R, Scilab,
Root, etc) 10000 exemplos de X e encontre computacionalmente as probabilidades
a seguir e compare com os valores teóricos esperados:
`a P(X < 1024).`
`b P(X < 1024|X > 961).`

```python
#Importando bibliotecas
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
media = 1000
desvio_padrao = 20

#amostras
n_amostras = 10000
#gerando variaveis aleatorias
amostras = np.random.normal(media, desvio_padrao, n_amostras)

# Calcular a probabilidade P(X < 1024)
probabilidade_a = norm.cdf(1024, loc=media, scale=desvio_padrao)
probabilidade_a
```

```python
# Plotar o gráfico para P(X < 1024)
plt.figure(figsize=(8, 5))
plt.hist(amostras, bins=30, density=True, alpha=0.5, color='blue', label='Amostras')
plt.axvline(x=1024, color='red', linestyle='--', label='X = 1024')
plt.legend()
plt.xlabel('X')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição de Probabilidade de X (P(X < 1024))')
plt.show()
```

```python
# Calcular a probabilidade P(X < 1024|X > 961)+
probabilidade_b = (norm.cdf(1024, loc=media, scale=desvio_padrao) - norm.cdf(961, loc=media, scale=desvio_padrao)) / (1 - norm.cdf(961, loc=media, scale=desvio_padrao))
```

```python
# Plotar o gráfico para P(X < 1024 | X > 961)
plt.figure(figsize=(8, 5))
plt.hist(amostras, bins=30, density=True, alpha=0.5, color='blue', label='Amostras')
plt.axvline(x=1024, color='red', linestyle='--', label='X = 1024')
plt.axvline(x=961, color='green', linestyle='--', label='X = 961')
plt.legend()
plt.xlabel(' X')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição de Probabilidade de X (P(X < 1024 | X > 961))')
plt.show()
```

```python
#valores teoricos esperados
print("Valor teórico de P(X < 1024):", probabilidade_a)
print("Valor teórico de P(X < 1024 | X > 961):", probabilidade_b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
tamanhos_amostra = [1000, 10000, 100000]
```

**Distribuição Uniforme para 1000**
--

```python
# Distribuição Uniforme

amostras_uniforme = np.random.uniform(0, 1, 1000)
# Histograma
plt.figure(figsize=(8, 4))
plt.hist(amostras_uniforme, bins=30, density=True, color='blue', alpha=0.7)
plt.title(f'Histograma - Distribuição Uniforme (1000 pontos)')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.show()
# Boxplot
plt.figure(figsize=(4, 4))
plt.boxplot(amostras_uniforme, vert=False)
plt.title(f'Boxplot - Distribuição Uniforme (1000 pontos)')
plt.show()

# Valor medio e variância
media = np.mean(amostras_uniforme)
variancia = np.var(amostras_uniforme)
variancia_teorica = 1/12
print(f'Distribuição Uniforme (1000 pontos):')
print(f'Média calculada: {media:.4f}, Média teórica: 0.5')
print(f'Variância calculada: {variancia:.4f}, Variância teórica: {variancia_teorica:.4f}')
```
